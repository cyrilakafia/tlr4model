## Creating .smi file for descriptor calculation

```python
import pandas as pd
import os
```

```python
df  = pd.read_csv("inactive_final.csv")
df
```

```python
# checking for duplicates across rows
# x = df[df.duplicated()]
# x

#checking for duplicates in specific columns
x = df[df.duplicated(['cid'])]
x
```

```python
df = df.drop(['cmpdname', 'activity'], axis = 'columns')
df
```

```python
smiles = df.pop('CanonicalSMILES')
df.insert(0, 'CanonicalSMILES', smiles)
df
```

```python
# creating a new directory in current directory for input and output descriptor files

os.mkdir('smi')
os.mkdir('des')
```

```python
# creating a .smi file

df.to_csv('smi/inactive.smi', sep = '\t', index = False, header = False)
```
