## Cleaning Inactives

The active dataset is cleaned and canonical smiles are extracted

```python
# importing libraries

import pandas as pd
import numpy as np
import pubchempy as pcp
import time
```

```python
# reading and importing active dataset

df = pd.read_csv("inactive.csv", low_memory= False)
df
```

```python
# checking for missing values in all columns

df.isnull().sum()
```

```python
# checking all column names

df.columns
```

```python
# removing 'unecessary' columns

df.drop(df.iloc[:, 2:24], inplace = True, axis = 1)
df
```

```python
cid_list = np.array(df.cid)

start = time.time()

smiles_df = pcp.get_properties(['CanonicalSMILES'], np.nditer(cid_list), 'cid', as_dataframe = True)

end = time.time()

print("The time take in seconds for extraction of canonical smiles is :  ", end-start)
smiles_df
```

```python
# removing indexes

smiles_df = smiles_df.reset_index(drop = True)
smiles_df
```

```python
#creating a csv of canonical smiles

smiles_df.to_csv("inactive_smiles.csv", index = False)
```

```python
# final datafram with all necessary features

final_df = pd.concat([df, smiles_df], axis = 1)
final_df
```

Rearranging features to put activity at the last column

```python
act = final_df.pop('activity')
final_df.insert(3, 'activity', act)
final_df
```

```python
# saving as a .csv file
final_df.to_csv("inactive_final.csv", index = False)
```
