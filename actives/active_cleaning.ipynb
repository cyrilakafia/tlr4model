## Cleaning Actives
The active dataset is cleaned and canonical smiles are extracted

```python
# importing libraries

import pandas as pd
import numpy as np
import pubchempy as pcp
import time
```

```python
# reading and importing active dataset

df = pd.read_csv("active.csv")
df
```

```python
# checking for missing values in all columns

df.isnull().sum()
```

```python
# checking all column names

df.columns
```

```python
# seperating the activity coloumn to add it later as the last column

act = df.activity
act_df = pd.DataFrame(act, columns = ['activity'])
act_df
```

```python
# removing 'unecessary' columns

df.drop(df.iloc[:, 2:25], inplace = True, axis = 1)
df
```

```python
# extracting canonical smiles from pubchem database

cid_list = np.array(df.cid)

start = time.time()

smiles_df = pcp.get_properties(['CanonicalSMILES'], np.nditer(cid_list), 'cid', as_dataframe = True)

end = time.time()

print("The time take in seconds for extraction of canonical smiles is :  ", end-start)
smiles_df
```

```python
# removing indexes

smiles_df = smiles_df.reset_index(drop = True)
smiles_df
```

```python
#creating a csv of canonical smiles

smiles_df.to_csv("active_smiles.csv", index = False)
```

```python
# final dataframe with all necessary features

final_df = pd.concat([df, smiles_df], axis = 1)
final_df
```

Rearranging features for to put activity at the last column

```python
# creating a new dataframe by adding act_df to fdf (final dataframe which excludes activity)

final_df = pd.concat([final_df, act_df], axis = 1)
final_df
```

Now activity column is the last

```python
# saving the final_df as a .csv file

final_df.to_csv("active_final.csv", index = False)
```
